```python
import spacy

# Load the spaCy model with word embeddings
nlp = spacy.load("en_core_web_md")

import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')
```

```python
def calculate_similarity(sentence1, sentence2):
    # Process the sentences with spaCy
    doc1 = nlp(sentence1)
    doc2 = nlp(sentence2)

    # Calculate the similarity between the sentences
    similarity = doc1.similarity(doc2)
    
    return similarity
```

```python
# Example usage
sentence1 = "The dog is chewing the bone"
sentence2 = "The dog is eating"
similarity_score = calculate_similarity(sentence1, sentence2)

print(f"Similarity between the sentences: {similarity_score}")
```

<h1>NLTK</h1>

```python
def preprocess_text(text):
    # Convert text to lowercase and remove punctuation
    text = text.lower()
    text = ''.join([char for char in text if char.isalnum() or char.isspace()])

    return text

def calculate_similarity(text1, text2):
    # Preprocess the text
    text1 = preprocess_text(text1)
    text2 = preprocess_text(text2)

    # Tokenize and remove stopwords
    stop_words = set(stopwords.words('english'))
    words1 = [word for word in nltk.word_tokenize(text1) if word not in stop_words]
    words2 = [word for word in nltk.word_tokenize(text2) if word not in stop_words]

    # Combine tokenized words back into sentences
    preprocessed_text1 = ' '.join(words1)
    preprocessed_text2 = ' '.join(words2)

    # Create TF-IDF vectorizer
    vectorizer = TfidfVectorizer()

    # Fit and transform the documents
    tfidf_matrix = vectorizer.fit_transform([preprocessed_text1, preprocessed_text2])

    # Calculate cosine similarity
    similarity = cosine_similarity(tfidf_matrix, tfidf_matrix)[0, 1]

    return similarity
```

```python
# Example usage
text1 = "The dog is chewing the bone"
text2 = "The dog is eating"
similarity_score = calculate_similarity(text1, text2)

print(f"Similarity between the texts: {similarity_score}")
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

def calculate_roberta_similarity(text1, text2):
    # Load the pre-trained RoBERTa model for sentence embeddings
    model = SentenceTransformer('roberta-base-nli-stsb-mean-tokens')

    # Encode the input texts into embeddings
    embeddings1 = model.encode([text1])
    embeddings2 = model.encode([text2])

    # Calculate cosine similarity between the embeddings
    similarity = cosine_similarity(embeddings1, embeddings2)[0, 0]

    return similarity

# Example usage
text1 = "The dog is chewing the bone"
text2 = "The dog is eating"
similarity_score = calculate_roberta_similarity(text1, text2)

print(f"RoBERTa Similarity between the texts: {similarity_score}")
```

```python
import numpy as np

# Simulated normalized similarity scores for illustration
similarity_scores = np.array([
    [0.816, 0.7, 0.9],  # spaCy
    [0.26, 0.6, 0.4],  # Jaccard
    [0.57, 0.8, 0.95]  # RoBERTa
])

# Step 1: Normalize the scores
normalized_scores = similarity_scores / np.linalg.norm(similarity_scores, axis=0)

# Step 2: Assign weights
weights = np.array([0.33, 0.33, 0.34])

# Step 3: Identify ideal and anti-ideal solutions
ideal_solution = np.max(normalized_scores, axis=1)
anti_ideal_solution = np.min(normalized_scores, axis=1)

# Step 4: Calculate Euclidean distances
distance_to_ideal = np.linalg.norm(normalized_scores - ideal_solution[:, np.newaxis], axis=1)
distance_to_anti_ideal = np.linalg.norm(normalized_scores - anti_ideal_solution[:, np.newaxis], axis=1)

# Step 5: Calculate similarity to ideal solution
similarity_to_ideal = distance_to_anti_ideal / (distance_to_ideal + distance_to_anti_ideal)

# Step 6: Rank the models
ranked_models = np.argsort(similarity_to_ideal)[::-1]

# Print the ranking
print("Model Ranking:")
for rank, model_idx in enumerate(ranked_models):
    print(f"{rank + 1}. Model {model_idx + 1} (Similarity Score: {similarity_to_ideal[model_idx]:.4f})")
```
